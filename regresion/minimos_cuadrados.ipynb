```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Establecemos una semilla para reproducibilidad
np.random.seed(42)

# Parámetros de la recta original
theta_0 = 52   # intercepto
theta_1 = 14   # pendiente

# Generamos 100 valores de x entre 0 y 10
X = np.linspace(0, 10, 100).reshape(-1, 1)

# Generamos y con la ecuación y = theta_0 + theta_1 * x
# Aumentamos significativamente el ruido (de 0.5 a 2)
Y = theta_0 + theta_1 * X + np.random.normal(0, 2, (100, 1))

# Creamos el modelo de regresión lineal
modelo = LinearRegression()

# Entrenamos el modelo (ajustamos el modelo a los datos)
modelo.fit(X, Y)

# Imprimimos los coeficientes estimados
print("Intercepto estimado (theta_0):", modelo.intercept_[0])
print("Pendiente estimada (theta_1):", modelo.coef_[0][0])

# Comparamos con los valores originales
print("\nValores originales:")
print(f"Theta_0 original: {theta_0}")
print(f"Theta_1 original: {theta_1}")

# Calculamos el coeficiente de determinación (R²)
r_squared = modelo.score(X, Y)
print(f"\nCoeficiente de Determinación (R²): {r_squared}")

# Graficamos los resultados
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Datos')
plt.plot(X, modelo.predict(X), color='red', label='Recta de Regresión')
plt.title('Regresión Lineal')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Set random seed for reproducible results
np.random.seed(42)

# Sample size
n_samples = 100

# Generate X values with normal distribution
# Center the data points around x=5 with some spread
x_values = np.random.normal(loc=5, scale=2, size=n_samples)
X = x_values.reshape(-1, 1)

# True parameters of the line
true_intercept = 2
true_slope = 1

# Add random noise to create y values
# This creates a normal distribution around the line
noise = np.random.normal(0, 0.5, n_samples)
Y = true_intercept + true_slope * X + noise.reshape(-1, 1)

# Fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Display model parameters
print(f"Fitted intercept: {model.intercept_[0]:.3f}")
print(f"Fitted slope:     {model.coef_[0][0]:.3f}")

# Calculate R-squared score
r_squared = model.score(X, Y)
print(f"R-squared score:  {r_squared:.3f}\n")

# Create scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Training data', alpha=0.6)
plt.plot(X, model.predict(X), color='red', label='Best fit line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
